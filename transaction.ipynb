```python
def calculate_netsaving(group):
    incoming = group.loc[group['Txn Amount'] > 0, 'Txn Amount'].sum()
    outgoing = group.loc[group['Txn Amount'] < 0, 'Txn Amount'].sum()
    return incoming + outgoing  
```

# AI Agent for Transaction Analysis

```python
from langchain_aws.chat_models.bedrock import ChatBedrock
from rich.console import Console
from rich.text import Text
import boto3
from langchain_experimental.agents.agent_toolkits import create_csv_agent
from langchain.agents.agent_types import AgentType
from langchain_aws import BedrockLLM
from langgraph.graph import StateGraph, START, END

from typing import TypedDict, Annotated
from langgraph.graph.message import add_messages


```

```python
region = 'us-east-1'
bedrock = boto3.client(
    service_name = 'bedrock-runtime',
    region_name = region,
    aws_access_key_id = 'AWS_ACCESS_KEY_ID',
    aws_secret_access_key = 'AWS_SECRET_ACCESS_KEY',
    aws_session_token = 'AWS_SESSION_TOKEN')
```

```python
model_kwargs =  {
    "max_tokens": 4000,
    "temperature": 1.0,
    "top_k": 250,
    "top_p": 0.999,
    "stop_sequences": ["\n\nhuman"],
}
```

```python
modelId = 'us.anthropic.claude-3-5-sonnet-20241022-v2:0'  



llm = ChatBedrock(
    model_id=modelId,
    client=bedrock,
    model_kwargs=model_kwargs,  
    beta_use_converse_api=True
)
```

```python
# Create a CSV agent

CSV_PATH = "transaction_data_1.csv"
agent = create_csv_agent(
    llm=llm,
    path=CSV_PATH,
    verbose=True,
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    handle_parsing_errors=True,  # Fixed typo: was "handlke_parsing_errors"
    allow_dangerous_code=True    # Required for security - allows code execution
)
```

```python
class State(TypedDict):
    df:pd.DataFrame
    question: Annotated[list, add_messages]
    answer: Annotated[list, add_messages]

graph = StateGraph(State)
```

```python
def load_csv(state: State):
    state['df'] = pd.read_csv(CSV_PATH)
    return {"df": state['df']}
```

```python
def query_agent(state:State):
    q = state["question"][-1]
    resp = agent.run(q)
    return {"answer": [resp]}
```

```python
graph.add_node("load_csv", load_csv)
graph.add_node("query_agent", query_agent)
graph.set_entry_point("load_csv")
graph.add_edge("load_csv","query_agent")
graph.set_finish_point("query_agent")

compiled = graph.compile()
```

```python
# Let's create a better version with proper JSON output parsing
import json
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate

# Define a more structured prompt for JSON output
json_prompt = """
Analyze the transaction data and provide insights for each user in the following JSON format:

{
  "analysis_summary": {
    "total_users": number,
    "analysis_date": "current_date"
  },
  "user_insights": {
    "USER_ID": {
      "spending_behavior": "description",
      "top_category": "category_name",
      "transaction_frequency": "frequency_description",
      "risk_profile": "low/medium/high",
      "recommended_offers": [
        {
          "type": "offer_type",
          "title": "offer_title",
          "description": "offer_description",
          "relevance_score": 0.8
        }
      ]
    }
  }
}

Analyze the CSV data and return ONLY the JSON response, no additional text.
Focus on users: US001, US002, US003, US004, US005
"""

# Create a new query function with better JSON handling
def query_agent_json(state: State):
    """Query agent with JSON output parsing"""
    q = state["question"][-1]
    
    try:
        # Get response from agent
        resp = agent.run(q)
        
        # Try to extract JSON from the response
        if isinstance(resp, str):
            # Look for JSON in the response
            start_idx = resp.find('{')
            end_idx = resp.rfind('}') + 1
            
            if start_idx != -1 and end_idx != 0:
                json_str = resp[start_idx:end_idx]
                try:
                    parsed_json = json.loads(json_str)
                    return {"answer": [parsed_json]}
                except json.JSONDecodeError:
                    # If JSON parsing fails, return structured text
                    return {"answer": [{"raw_response": resp, "status": "json_parse_failed"}]}
            else:
                return {"answer": [{"raw_response": resp, "status": "no_json_found"}]}
        else:
            return {"answer": [resp]}
            
    except Exception as e:
        return {"answer": [{"error": str(e), "status": "agent_error"}]}

print("✅ Improved JSON query function created!")
```

```python
# Alternative: Simple user-by-user analysis (more reliable)
def analyze_users_individually():
    """Analyze each user individually for more reliable results"""
    
    # Load the data first
    df = pd.read_csv(CSV_PATH)
    users = df['User Id'].unique()
    
    results = {}
    
    print(f"Analyzing {len(users)} users individually...")
    
    for user in users:
        print(f"\n Analyzing user: {user}")
        
        # Create focused prompt for this user
        user_prompt = f"""
        Analyze user {user} from the transaction data and provide:
        1. Total spending amount
        2. Top spending category
        3. Spending behavior (conservative/moderate/high spender)
        4. Recommended offers (2-3 specific offers)
        
        Keep the response concise and focused.
        """
        
        try:
            # Query the agent for this specific user
            response = agent.run(user_prompt)
            results[user] = response
            print(f"Completed analysis for {user}")
            
        except Exception as e:
            print(f" Error analyzing {user}: {e}")
            results[user] = f"Error: {e}"
    
    return results

# Run the individual analysis
print("Starting individual user analysis...")
individual_results = analyze_users_individually()

# Display results
print("\n" + "="*80)
print("INDIVIDUAL USER ANALYSIS RESULTS")
print("="*80)

for user, analysis in individual_results.items():
    print(f"\n USER: {user}")
    print("-" * 40)
    print(analysis)
    print("-" * 40)
```
